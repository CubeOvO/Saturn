```python
import pandas as pd
import numpy as sp
import numpy as np
import matplotlib.pyplot as plt

# show plots without need of calling `.show()`
%matplotlib inline
```

```python
df = pd.read_csv('-veioncsc.csv')
# df[str(0)]
a = str(0);a
```

```python
1# calculate the width for each bin

w = df['Wavelength']
f = df['Flux']



def width(w):
    d_w = []
    for i in range(len(w)):
        if i < len(w)-1:
            d_w.append(w[i+1]-w[i])
    return d_w

```

```python
df1 = pd.read_csv('hcs2.csv')
df1
```

```python
sw = w[17:283]
sf = f[17:283]
cw = df1['w']
cs = df1['cs']
```

```python
len(sf)
```

```python
cw_c = []
for i in range(len(sf)+2):
#     print(i)
    found = False
    for j in range(len(cw)):
#         print(j)
        if cw[j]>=sw[i+17] and cw[j]<=sw[i+17]:
            cw_c.append(cs[j])
            found = True
            print(cs[j])
            break
    if not found:
        for j in range(len(cw)-1):
            if j>0:
                if cw[j]>=sw[i+17]:
                    cw_c.append(cs[j-1])
                    found = True
                    break
            else:
                if cw[j]>=sw[i+17]:
                    cw_c.append(cs[j])
                    found = True
                    break
    if not found:
        cw_c.append(0)
        print('error')
        break
```

```python
dw = width(w)
print(dw)
dw = dw[17:283]
print(dw)
```

```python
k_rate = []
height = []
for i in range(len(sw)-1):
    k = cw_c[i]*list(sf)[i]
    k_rate.append(k*(10**-18))
    height.append(k*(10**-18)/dw[i])
plt.yscale('log')
plt.xlim([0,20000])
plt.plot(sw[:-1],height)
sum(k_rate) 
```
