```python
import pandas as pd
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt

# show plots without need of calling `.show()`
%matplotlib inline
```

```python
df = pd.read_csv('data.csv')
df
```

```python
a = df['Flux']
```

```python
a = a.astype(np.float)
```

```python
params = {
   'axes.labelsize': 18,
   'font.size': 8,
   'legend.fontsize': 18,
   'xtick.labelsize': 13,
   'ytick.labelsize': 13,
   'figure.figsize': [8,8]
   } 
plt.rcParams.update(params)

plt.xlabel('LOG WAVELENGTH (A)')
plt.ylabel('LOG PHOTONS (cm-2 s-1 A-1)')

plt.loglog(df['Wavelength'],a)
# plt.loglog(df['Wavelength'],a,'.')
```
